## Question 5: Plot Tesla Stock Graph


Use the `make_graph` function to graph the Tesla Stock Data, also provide a title for the graph. Note the graph will only show data upto June 2021.


<details><summary>Hint</summary>

```

You just need to invoke the make_graph function with the required parameter to print the graphs.The structure to call the `make_graph` function is `make_graph(tesla_data, tesla_revenue, 'Tesla')`.

```
    
</details>


````python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

def make_graph(stock_data, revenue_data, stock):
    # Create subplots: 2 rows, 1 column
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price", "Historical Revenue"), vertical_spacing = .3)

    # Plot the stock data (historical share price)
    stock_data_specific = stock_data[stock_data['Date'] <= '2021-06-14']
    fig.add_trace(go.Scatter(x=stock_data_specific['Date'], y=stock_data_specific['Close'], name='Stock Price', line=dict(color='blue')), row=1, col=1)

    # Plot the revenue data (historical revenue)
    revenue_data_specific = revenue_data[revenue_data['Date'] <= '2021-04-30']
    fig.add_trace(go.Bar(x=revenue_data_specific['Date'], y=revenue_data_specific['Revenue'], name='Revenue', marker=dict(color='orange')), row=2, col=1)

    # Update layout
    fig.update_layout(
        title=f'{stock} Stock and Revenue Data',
        xaxis_title="Date",
        yaxis_title="Price (USD)",
        yaxis2_title="Revenue (Millions)",
        showlegend=True
    )

    fig.show()

# Now you can use the make_graph function to plot your data
make_graph(tesla_data, tesla_revenue, 'Tesla')
````

## Question 6: Plot GameStop Stock Graph


Use the `make_graph` function to graph the GameStop Stock Data, also provide a title for the graph. The structure to call the `make_graph` function is `make_graph(gme_data, gme_revenue, 'GameStop')`. Note the graph will only show data upto June 2021.


<details><summary>Hint</summary>

```

You just need to invoke the make_graph function with the required parameter to print the graphs.The structure to call the `make_graph` function is `make_graph(gme_data, gme_revenue, 'GameStop')`

```
    
</details>


````python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def make_graph(stock_data, revenue_data, stock):
    # If 'Date' is the index, use it directly or reset the index to make it a column
    stock_data = stock_data.reset_index()  # This makes 'Date' a column

    # Ensure the 'Date' column is in datetime format
    stock_data['Date'] = pd.to_datetime(stock_data['Date'])
    revenue_data['Date'] = pd.to_datetime(revenue_data['Date'])

    # Check if the data is not empty and has the correct structure
    print(stock_data.head())
    print(revenue_data.head())

    # Create subplots: 2 rows, 1 column
    fig = make_subplots(
        rows=2, cols=1, 
        shared_xaxes=True, 
        subplot_titles=(f"{stock} Stock Price", f"{stock} Revenue"), 
        vertical_spacing = .3
    )

    # Plot the stock data (historical share price) up to June 2021
    stock_data_specific = stock_data[stock_data['Date'] <= '2021-06-01']
    fig.add_trace(go.Scatter(x=stock_data_specific['Date'], y=stock_data_specific['Close'], 
                             name=f'{stock} Stock Price', line=dict(color='blue')), row=1, col=1)

    # Plot the revenue data (historical revenue) up to June 2021
    revenue_data_specific = revenue_data[revenue_data['Date'] <= '2021-06-01']
    fig.add_trace(go.Bar(x=revenue_data_specific['Date'], y=revenue_data_specific['Revenue'], 
                         name=f'{stock} Revenue', marker=dict(color='orange')), row=2, col=1)

    # Update layout
    fig.update_layout(
        title=f'{stock} Stock and Revenue Data (up to June 2021)',
        xaxis_title="Date",
        yaxis_title="Price (USD)",
        yaxis2_title="Revenue (Millions)",
        showlegend=True
    )

    # Show the plot
    fig.show()

# Example usage with stock data and revenue data
# Ensure to replace 'gme_data' and 'gme_revenue' with actual data
make_graph(gme_data, gme_revenue, 'GameStop')
````
